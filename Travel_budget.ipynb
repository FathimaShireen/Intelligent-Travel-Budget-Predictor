```python
from sklearn.preprocessing import LabelEncoder
le_city = LabelEncoder()
le_season = LabelEncoder()
le_flight = LabelEncoder()
le_hotel = LabelEncoder()
le_transport = LabelEncoder()
le_cost = LabelEncoder()  

df["City"] = le_city.fit_transform(df["City"].astype(str))
df["Season"] = le_season.fit_transform(df["Season"].astype(str))
df["Flight_Type"] = le_flight.fit_transform(df["Flight_Type"].astype(str))
df["Hotel_Class"] = le_hotel.fit_transform(df["Hotel_Class"].astype(str))
df["Transport_Mode"] = le_transport.fit_transform(df["Transport_Mode"].astype(str))
df["Cost_Category"] = le_cost.fit_transform(df["Cost_Category"].astype(str))
```

```python
x=df.drop(['Total_Trip_Cost','Cost_Category'],axis=1).values
x
```

```python
y=df[['Total_Trip_Cost','Cost_Category']]
y
```

```python
y_reg = y['Total_Trip_Cost']
y_reg
```

```python
y_class=y['Cost_Category']
y_class
```

```python
print(y_class.value_counts())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_reg_train,y_reg_test,y_class_train,y_class_test=train_test_split(x,y_reg,y_class,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(x_train, y_class_train)

print("\nClass distribution AFTER SMOTE:")
print(pd.Series(y_train_res).value_counts())
```

```python
y_train_res
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.naive_bayes import BernoulliNB
base=BernoulliNB()
from sklearn.svm import SVC
sv=SVC()
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy')
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(criterion='entropy')
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
lst=[knn,base,sv,tree,forest]
lst1=[]


for i in lst:
    print('Classification Model is : ',i)
    i.fit(x_train,y_class_train)
    y_pred=i.predict(x_test)
    cm=confusion_matrix(y_class_test,y_pred)
    print('Confusion Matrix')
    print(cm)
    score=accuracy_score(y_class_test,y_pred)
    print('Accuracy Score : ',score)
    report=classification_report(y_class_test,y_pred)
    print('Classification Report')
    print(report)
    lst1.append(score)
    print('********************************************************************************')
```

```python
from sklearn.tree import DecisionTreeRegressor
tree1=DecisionTreeRegressor(criterion='squared_error')
from sklearn.ensemble import RandomForestRegressor
forest1=RandomForestRegressor(criterion='squared_error')
from sklearn.metrics import mean_absolute_percentage_error,r2_score
lst=[tree1,forest1]
lst1=[]

for i in lst:
    print('Regression Model is : ',i)
    i.fit(x_train,y_reg_train)
    y_pred=i.predict(x_test)
    pct=mean_absolute_percentage_error(y_reg_test,y_pred)
    print('Percentage Error : ',pct)
    score2=r2_score(y_reg_test,y_pred)
    print('Score : ',score2)
    lst1.append(score2)
    print('***********************************************************')
```

```python
import joblib
reg_model=forest1
clf_model=forest
```

```python
joblib.dump(reg_model,"reg_model.pkl")   # Regression model (Total_Trip_Cost)
joblib.dump(clf_model,"clf_model.pkl")   # Classification model (Cost_Category)
joblib.dump(scaler,"scaler.pkl")
```

```python
#Which city offers the cheapest travel experience overall?

df["City_Name"] = le_city.inverse_transform(df["City"])

sns.barplot(data=df, x="City_Name", y="Total_Trip_Cost",color='r')
plt.title("Average Total Trip Cost by City")
plt.tight_layout()
plt.show()
```

```python
#In which season do most budget trips occur?

df["Season_Name"] = le_season.inverse_transform(df["Season"])
df["Cost_Label"] = le_cost.inverse_transform(df["Cost_Category"])

budget_season= (df[df["Cost_Label"] == "Budget"]["Season_Name"].value_counts().reset_index(name="Budget_Trip"))

sns.barplot(data=budget_season,x="Season_Name",y="Budget_Trip",color='g')
plt.title("Season with Highest Budget Trips")
plt.xlabel("Season")
plt.ylabel("Budget Trip Count")
plt.tight_layout()
plt.show()
```
